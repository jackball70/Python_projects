```python
import os 
import pandas as pd 
import matplotlib.pyplot as mp  
import numpy as np  
import seaborn as sns 

```

```python
path = 'C:/NFL_data/'    
extension = '.csv'  

files = [file for file in os.listdir(path) if file.endswith(extension)]

playerdf = pd.DataFrame 
teamdf = pd.DataFrame

for file in files: 
    if 'player' in file:  
        playerdf = pd.read_csv(os.path.join(path, file))  
    else:
        teamdf = pd.read_csv(os.path.join(path, file)) 
    
```

```python
playerdf = playerdf.loc[playerdf['season'] >= 2019] 
playerdf = playerdf.loc[playerdf['position'] == 'WR']  



playerdf.head()
```

```python


wrfeats = ['player_id', 'player_name', 'season', 'targets', 'receptions', 'receiving_yards', 'receiving_air_yards', 'yards_after_catch', 'target_share', 'air_yards_share', 'total_tds']  

groupby_feats = ['player_id', 'player_name', 'season']


wrdf = (playerdf.loc[:, wrfeats].groupby(groupby_feats, as_index= False).sum()) 

wrdf.sample(10)
```

```python
for y in ['targets', 'receptions', 'receiving_yards', 'receiving_air_yards',  
          'yards_after_catch', 'target_share', 'air_yards_share']: 
  sns.regplot(data=wrdf, x='total_tds', y=y) 
  mp.title(f"touchdowns and {y}") 
  mp.show()
```

```python
newdf = wrdf.copy()  

newdf['season'] = newdf['season'].add(1) 

new_wrdf = (wrdf.merge(newdf, on=['season', 'player_name', 'player_id'], suffixes=('', '_prev'), how='left')) 
new_wrdf.sample(10)
```

```python
for y in ['targets_prev', 'receptions_prev', 'receiving_yards_prev', 'receiving_air_yards_prev',  
          'yards_after_catch_prev', 'target_share_prev', 'air_yards_share_prev']: 
  sns.regplot(data=new_wrdf, x='total_tds', y=y) 
  mp.title(f"touchdowns and {y}") 
  mp.show()
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error 
from scipy.stats import pearsonr
```

```python
new_wrdf.sample(10)
```

```python
features = ['targets_prev', 'receptions_prev', 'receiving_yards_prev', 'receiving_air_yards_prev',  
          'yards_after_catch_prev', 'target_share_prev', 'air_yards_share_prev'] 

target = 'total_tds' 

model_data = (new_wrdf.dropna(subset=features+[target])) 

train_data = (model_data.loc[model_data['season']==2021]) 

test_data = (model_data.loc[model_data['season']==2022]) 

model = LinearRegression() 

model.fit(train_data.loc[:, features], train_data[target])  

preds = model.predict(test_data.loc[:, features])

preds = pd.Series(preds, index=test_data.index) 

test_data['preds'] = preds
```

```python
rmse = mean_squared_error(test_data['total_tds'], test_data['preds'])**0.5 
r2 = pearsonr(test_data['total_tds'], test_data['preds'])[0]**2 
print(f"rmse is equal to: {rmse}, \nr2 is equal to: {r2}")
```

```python
sns.regplot(data=test_data, x='total_tds', y='preds') 
mp.title('Touchdowns and Predictions') 
mp.show()
```
